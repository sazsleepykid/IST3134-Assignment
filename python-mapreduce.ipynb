```python
import numpy as np
import pandas as pd
from collections import Counter
import re
import time

# Define the file path and read the data
csv_file_path = 'C:/2/steam.review.csv'
data = pd.read_csv(csv_file_path, encoding='utf-8')

# Function to preprocess the text
def preprocess_text(text):
    # Check if the text is a valid string (not NaN)
    if isinstance(text, str):
        # Convert to lowercase
        text = text.lower()
        # Remove non-alphanumeric characters
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
        return text
    else:
        # Return an empty string for NaN values
        return ''

# Apply preprocessing to the review text
data['clean_text'] = data['review_text'].apply(preprocess_text)

# Apply preprocessing to the review text
data['clean_text'] = data['review_text'].apply(preprocess_text)

# Define the word count function
def word_count(text):
    words = text.split()
    return len(words)

# Record start time
start_time = time.time()

# Apply word count to the clean text
data['word_count'] = data['clean_text'].apply(word_count)

# Record end time
end_time = time.time()

# Calculate total processing time
total_processing_time = end_time - start_time
print("Total Processing Time:", total_processing_time, "seconds")

# Get the total word count
total_word_count = data['word_count'].sum()
print("Total Word Count:", total_word_count)
```
